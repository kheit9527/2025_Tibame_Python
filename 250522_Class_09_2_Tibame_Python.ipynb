```
HTTP代碼：
1. 2開頭: 成功
2. 3開頭: 轉址
3. 4開頭: 錯誤
** 404: Not Found(找不到)
** 403: Forbidden(禁止)
    1. IP Ban
    ** 第一次就被拒絕不是你人(IP)有問題，而是有其他問題
    2. 缺少了一些很重要的headers
    ** 瀏覽器的網址一模一樣的時候，它根本不能知道你是"點擊"還是"跑程式"
    ** 所以只有可能：你送的跟瀏覽器不一樣(你缺少了headers)
    ** 你的電腦網址(url)+headers 送過去 -> 伺服器回應你response+headers
    ** headers -> user-agent: 所有東西都是你告訴它(伺服器)的
    ** accept-language -> 告訴你語言的網頁
       (所以不同國家電腦，key同一網址，會有不同語言的結果)

** DDOS: 上萬台殭屍電腦攻擊拜訪一個網址(無法拒絕，只能找更大流量網址確認是正常/不正常)
```

````python
import urllib.request as req
import bs4

url = "https://www.ptt.cc/bbs/Beauty/M.1747794743.A.FA0.html"
dn = url.split("/")[-1]
# dn = "/content/drive/MyDrive/ptt" + url.split("/")[-1]
if not os.path.exists(dn):
    os.makedirs(dn)

r = req.Request(url)
r.add_header("user-agent", "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/136.0.0.0 Safari/537.36 Edg/136.0.0.0")

resp = req.urlopen(r)
content = resp.read()

html = bs4.BeautifulSoup(content)

allow_subs = ["jpg", "jpeg", "png", "gif"]
links = html.find_all("a")
for link in links:
    link_href = link["href"]
    sub = link_href.split(".")[-1]
    if sub.lower() in allow_subs: # 不區分大小寫: 全部都轉小寫
        img_url = link_href
        img_r = req.Request(url)
        img_r.add_header("user-agent", "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/136.0.0.0 Safari/537.36 Edg/136.0.0.0")

        img_resp = req.urlopen(img_r)
        img = img_resp.read()

        fn = img_url.split("/")[-1]
        fn = dn + "/" + fn
        # !!! 純文字: "w"/"r" + encoding="utf-8"
        # !!! 非純文字: "wb"/"rb"
        f = open(fn, "wb")
        f.write(img)
        f.close()

        !zip -r /content/file.zip /content/M.1747794743.A.FA0.html
````
