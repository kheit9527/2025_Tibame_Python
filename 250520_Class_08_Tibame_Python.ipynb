````python
# 從上次Class07_2繼續

import urllib.request as req
# beautifulsoup4
import bs4 as bs

url = "https://tabelog.com/tw/osaka/rstLst/sweets/?SrtT=rt"
resp = req.urlopen(url)
content = resp.read()
# print(content)
# json.loads

# html型態轉換: string字串 -> 區塊型態
html = bs.BeautifulSoup(content)

# 一層一層拆解 ex.我房間衣櫃的第三個抽屜裡的...
# class用一個就好
rlist = html.find_all("div",{"class":"list-rst__body"})

# list(清單)不能再繼續find, find是區塊的專屬功能
# 1. for ... in ... 2. [] 3.[-1]
for r in rlist:
    # print(type(r))
    en = r.find("a", {"class":"list-rst__rst-name-target"})
    en_text = en.text
    en_href = en["href"]
    print(en_text)
    print(en_href)
    print("-" * 30)
````

````python
# 練習1

import urllib.request as req
# beautifulsoup4
import bs4 as bs

url = "https://tabelog.com/tw/osaka/rstLst/sweets/?SrtT=rt"
resp = req.urlopen(url)
content = resp.read()
# print(content)
# json.loads

# html型態轉換: string字串 -> 區塊型態
html = bs.BeautifulSoup(content)

# 1.評價
rlist = html.find_all("div",{"class":"list-rst__body"})
# print(rlist)

for p in rlist:
    rate_val = p.find("span", {"class":"c-rating__val"})
    rate_val_text = rate_val.text
    print(rate_val_text)

# 2.晚間價錢
for q in rlist:
    dinner_list = q.find("span", {"class":"c-rating-v3__val"})
    dinner_text = dinner_list.text
    print(dinner_text)

# 3.午間價錢
for r in rlist:
    lunch_price = r.find("span", {"class":"c-rating-v3__val"})
    lunch_text = lunch_price.text
    print(lunch_text)

# 4.哪天沒開
for s in rlist:
    closed_day = s.find("span", {"class":"list-rst__holiday-text"})
    closed_day_text = open_day.text
    print(closed_day_text)

# 5. 所在地區
for t in rlist:
    location_stamp = t.find("div", {"class":"list-rst__area-genre"})
    location_stamp_text = location_stamp.text
    print(location_stamp_text)
````

````python
# 老師寫的3

import urllib.request as req
# beautifulsoup4
import bs4 as bs
# 0.
import pandas as pd

i = 0

# 1. 先準備一個空的list
table = []

for i in range(5):
    page = i + 1
    url = "https://tabelog.com/tw/osaka/rstLst/sweets/" + str(page) + "/?SrtT=rt"
    print(url)
    resp = req.urlopen(url)
    content = resp.read()
    # print(content)
    # json.loads

    # html型態轉換: string字串 -> 區塊型態
    html = bs.BeautifulSoup(content)

    rlist = html.find_all("div",{"class":"list-rst__body"})

    # 評價 / 晚餐價格 / 午餐價格 / 哪天沒開 / 地點
    for r in rlist:
        en = r.find("a", {"class":"list-rst__rst-name-target"})
        rating = r.find("span", {"class":"c-rating__val"})
        prices = r.find_all("span", {"class":"c-rating-v3__val"})
        dinner_price = prices[0]
        lunch_price = prices[1]
        holiday = r.find("span", {"class":"list-rst__holiday-text"})
        area_genre = r.find("div", {"class":"list-rst__area-genre"})
        message = r.find("em", {"class":"list-rst__rvw-count-num"})


        en_text = en.text
        en_href = en["href"]
        rating_text = rating.text
        dinner_text = dinner_price.text
        lunch_text = lunch_price.text
        holiday_text = holiday.text
        area_genre_text = area_genre.text
        area_genre_text = area_genre_text.strip()
        message_text = message.text

        print(en_text)
        print(en_href)
        print(rating_text)
        print(dinner_text)
        print(lunch_text)
        print(holiday_text)
        print(area_genre_text)
        print(f"留言數量: {message_text}")
        print("-" * 30)

        # 2. 拿到一筆就把字典準備好
        data = {
        "en": en_text,
        "en_href": en_href,
        "rating": rating_text,
        "dinner": dinner_text,
        "lunch": lunch_text,
        "holiday": holiday_text,
        "area_genre": area_genre_text,
        "message": message_text
        }

        # 3. append
        table.append(data)
        i = i + 1

# 4. JSON -> DataFrame
df = pd.DataFrame(table)
df.to_csv("tabelog.csv", encoding="utf-8")
````
